```python
import numpy as np
from scipy import fftpack
import matplotlib.pyplot as plt
%matplotlib inline
```

# Create a Signal

```python
t = np.linspace(0.,1e-1,1000) # time vector
dt = (t.max() - t.min())/float(t.size) # timestep

f = 200. #base frequency of signal
phi = 0. # phase shift (rad)

signal = np.sin(2.*np.pi*f*t+phi)
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,3))
ax.plot(t,signal)
ax.set_xlim([t.min(), t.max()])
ax.set_title('TimeSeries')
```

# Take the FFT

```python
# Take FFT
Signal = fftpack.fft(signal)
freq = fftpack.fftfreq(t.size,dt)

# FFT shift
Signal = fftpack.fftshift(Signal)
freq = fftpack.fftshift(freq)
```

```python
fig, ax = plt.subplots(3,1,figsize=(8,9))

ax[0].plot(freq,Signal.real)
ax[0].set_title('Real')

ax[1].plot(freq,Signal.imag)
ax[1].set_title('Imag')

ax[2].plot(freq,np.abs(Signal))
ax[2].set_title('Abs')

plt.tight_layout()
```
