```python
def Response(alpha):
    return (alpha**2+1j*alpha) / (1+alpha**2)
```

```python
alpha = np.logspace(-3,3, 61)
```

```python
import matplotlib
```

```python
matplotlib.rcParams['font.size'] = 18
matplotlib.rcParams['text.usetex'] = True
```

```python
fig = plt.figure(figsize = (8,5))
plt.semilogx(alpha, Response(alpha).real, 'k', lw=2)
plt.semilogx(alpha, Response(alpha).imag, 'r', lw=2)
plt.xlabel("$\\alpha$", fontsize = 25)
plt.ylabel("$Q(\\alpha$)", fontsize = 25)
plt.grid(True)
plt.legend(("Real", "Imaginary"), fontsize = 16)
fig.savefig("CmplxResp.png")
```
