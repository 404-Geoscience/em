```python
def ColeColePelton(f, sigmaInf, eta, tau, c, option):
    """
    
        .. math::
            \sigma(\omega) = \sigma_{\infty}\Big(1-\frac{\eta}{1+(1-\eta)(\imath\omega\tau)^c} \Big)
            
    """
    w = 2*np.pi*f
    sigma = sigmaInf*(1 - eta/(1 + (1-eta)*(1j*w*tau)**c))
    if option=="sigma":
        return sigma
    elif option=="resis":
        return 1./(sigma)
    else:
        raise Exception("Put only sigma or resis")
```
